```python
import os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

# mpl.rc('font', family='Malgun Gothic')
```

```python
vehicle = pd.read_csv('/Users/isc-yeon/Documents/projectData/Vehicle_Registration_Transactions_by_Department_of_Licensing.csv',
                      # header = 0,
                      # index_col = 'Model Year',
                      # squeeze = True,
                      # nrows=10_000,
                      usecols = [3,11],
                      low_memory=False)

vehicle
```

```python
df = (
    vehicle
    .groupby(['Model Year', 'Electrification Level'])
    .agg(size_Electrification_Level=pd.NamedAgg(column='Electrification Level', aggfunc='size'),
        ).reset_index()
)
df
```

```python
df['Datetime'] = pd.to_datetime(df['Model Year'], format='%Y')
df
```

```python
EV = df[df['Electrification Level'] == 'ICE (Internal Combustion Engine)'].index
EV
```

```python
df_EV = df.drop(EV).reset_index(drop=True)
df_EV
```

```python
df_EV.loc[:,'Electrification Level'] = 'EV'
```

```python
df_EV
```

```python
import statsmodels.tsa.api as tsa
```

```python
df_time= pd.pivot_table(data=df_EV, index='Datetime', 
                        values='size_Electrification_Level', aggfunc='sum')
df_time
```

```python
df_time_Del23 = df_time.drop('2023-01-01')
```

```python
y = df_time['size_Electrification_Level'].resample('1Y').sum()
y_Del23 = df_time_Del23['size_Electrification_Level'].resample('1Y').sum()
y
```

```python
y.isnull().sum()
```

```python
from pylab import rcParams

rcParams['figure.figsize'] = 15, 10
```

```python
# 시계열 모델 생성
model_series = tsa.seasonal_decompose(y, model='additive')
# 모델 시각화
fig = model_series.plot()
plt.show()
```

```python
import itertools
```

```python
p = d = q = range(0, 2)
pdq = list(itertools.product(p, d, q))
seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]
```

```python
seasonal_pdq
```

```python
param_list = []
param_seasonal_list = []
results_AIC_list = []
```

```python
for param in pdq:
     for param_seasonal in seasonal_pdq:
        try:
            mod = tsa.statespace.SARIMAX(y,order=param,
                                            seasonal_order=param_seasonal,
                                            enforce_stationarity=False,
                                            enforce_invertibility=False)
            results = mod.fit()
            param_list.append(param)
            param_seasonal_list.append(param_seasonal)
            results_AIC_list.append(results.aic)
        except:
             continue
```

```python
ARIMA_list = pd.DataFrame({'Parameter':param_list,'Seasonal':param_seasonal_list,'AIC':results_AIC_list})
```

```python
ARIMA_list.sort_values(by='AIC')
```

```python
mod = tsa.statespace.SARIMAX(y,order=(0, 1, 1),seasonal_order=(0, 1, 1, 12),
                              enforce_stationarity=False, enforce_invertibility=False)
results = mod.fit()
print(results.summary())
```

```python
results.plot_diagnostics(figsize=(16, 8))
plt.show()
```

```python
results.get_prediction()
```

```python
y.head()
```

```python
pred = results.get_prediction(start=pd.to_datetime('1981-12-31'), dynamic=False)
# pred = results.get_prediction(start=pd.to_datetime('1966-12-31'), dynamic=False)
pred_ci = pred.conf_int()

ax = y.plot(label='observed')
pred.predicted_mean.plot(ax=ax, label='One-step ahead Forecast', alpha=.7, figsize=(14, 7))
ax.fill_between(pred_ci.index,pred_ci.iloc[:, 0],pred_ci.iloc[:, 1], color='k', alpha=.2)

plt.legend()
plt.show()
```

```python
pred = results.get_prediction(start=pd.to_datetime('1981-12-31'), dynamic=False)
# pred = results.get_prediction(start=pd.to_datetime('1966-12-31'), dynamic=False)
pd.DataFrame(pred.predicted_mean).reset_index()
```

```python
y_forecasted = pred.predicted_mean
y_truth = y['1981-01-01':]
# y_truth = y['1966-01-01':]
mse = ((y_forecasted - y_truth) ** 2).mean()
print('MSE {}'.format(round(mse, 2)))
```
